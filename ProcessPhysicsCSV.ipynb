```python
import pandas as pd
import plotly.graph_objects as go
```

```python
df = pd.read_csv('.\Data\PhysicsSuite\WalkingLevel\WalkingLevel_Phys_1.csv')
df.index = pd.to_datetime(df['time'], unit = 's')
df = df[df['time']<=16]
```

```python
fig = go.Figure()
for axis in ['gFx', 'gFy', 'gFz']:
    fig.add_trace(go.Scatter(x = df.index, y = df[axis], name = axis))
    
fig.update_layout(title=go.layout.Title(text="Gravitional Force", font=dict(
                family="Courier New, monospace",
                size=22,
                color="#0000FF"
            )))
fig.show()
fig = go.Figure()

for axis in ['ax', 'ay', 'az']:
    fig.add_trace(go.Scatter(x = df.index, y = df[axis], name = axis))
fig.update_layout(title=go.layout.Title(text="Acceleration", font=dict(
                family="Courier New, monospace",
                size=22,
                color="#0000FF"
            )))
fig.show()

fig = go.Figure()
for axis in ['wx', 'wy', 'wz']:
    fig.add_trace(go.Scatter(x = df.index, y = df[axis], name = axis))
fig.update_layout(title=go.layout.Title(text="Gyro", font=dict(
                family="Courier New, monospace",
                size=22,
                color="#0000FF"
            )))
fig.show()

fig = go.Figure()
for axis in ['Azimuth', 'Pitch', 'Roll']:
    fig.add_trace(go.Scatter(x = df.index, y = df[axis], name = axis))
fig.update_layout(title=go.layout.Title(text="Heading", font=dict(
                family="Courier New, monospace",
                size=22,
                color="#0000FF"
            )))
fig.show()
```
